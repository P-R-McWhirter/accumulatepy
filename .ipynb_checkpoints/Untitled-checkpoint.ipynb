```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import os
import csv
import glob
from shutil import copyfile

## Open a list of file names in the current work directory cwd.

cwd = "G:\Heimdall\EWT_Camera Traps\Bijstein"

out = "G:\Heimdall\EWT_Camera Traps\Training"

files = glob.glob(cwd + '/**/*.jpg', recursive=True)

## For each of the images, save the file paths to argument name.

for i in files:
    i_n = i[38:]
    i_n = i_n.replace("\\", "-")
    i_n = i_n.replace(" ", "_")
    output = out + "\\" + i_n
    copyfile(i, output)
```
